# **Algoritmo de Deutsch**

- Soluciona el problema de paridad para el caso donde $n=1$.
- Problema en computación cuántica usualmente referido como *Problema de Deutsch*.

- Función: $f: \sum \to \sum$

|$$a$$|$$f_1(a)$$|$$f_2(a)$$|$$f_3(a)$$|$$f_4(a)$$| 
| -- | -- | -- | -- | -- |
| 0 | 0 | 0 | 1 | 1 |
| 1 | 0 | 1 | 0 | 1 |

- La primera y última de las funciones son constantes. Las del medio son balanceadas. 
- El problema de Deutsch consiste en saber a cual de las categorías pertenece el input.

**Problema de Deutsch**
- Input: función $f: \{0,1\} \to \{0,1\}$
- Output: 0 si $f$ es constante, 1 si $f$ es balanceada
